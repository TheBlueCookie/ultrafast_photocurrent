# MoS$_2$-MoSe$_2$ hetereostructure circuit simulation

```python
import PySpice.Logging.Logging as Logging
from PySpice.Spice.Netlist import Circuit
from PySpice.Unit import *
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from datetime import datetime

logger = Logging.setup_logging()
```

```python
def sweep_param(mode, sweeprange, count, current_source, my_simulator, elements, log=True, output=False, choppermode=False):
    modes = ['capacitance', 'resistance', 'frequency', 'inductance']
    assert mode in modes, f'{mode} cannot be varied.'
    
    start, stop = sweeprange
    if mode == 'frequency':
        start = 1/(2*start)
        stop = 1/(2*stop)
        if choppermode:
            start *= 1/2
            stop *= 1/2
    
    if not log:
        sweep = np.linspace(start, stop, count)
    else:
        sweep = np.logspace(np.log10(start), np.log10(stop), count)
        
    res = []
    
    if output:
        print(sweep)
        
    if mode == 'resistance':
        for val in sweep:
            for el in elements:
                el.resistance = val
            res.append((my_simulator.transient(step_time=current_source.period/resolution, 
                                              end_time=current_source.period*length), val))

    elif mode == 'capacitance':
        for val in sweep:
            for el in elements:
                el.capacitance = val
            res.append((my_simulator.transient(step_time=current_source.period/resolution, 
                                              end_time=current_source.period*length), val))
    elif mode == 'inductance':
        for val in sweep:
            for el in elements:
                el.inductance = val
            res.append((my_simulator.transient(step_time=current_source.period/resolution, 
                                              end_time=current_source.period*length), val))
    
    elif mode == 'frequency':
        for val in sweep:
            current_source.pulse_width = val
            current_source.period = 2*val
            res.append((my_simulator.transient(step_time=current_source.period/resolution, 
                                              end_time=current_source.period*length), 1/current_source.period))
    
    return res
```

```python
# initial circuit parameters

# circuit parameters
junction_resistance = 100@u_kΩ
junction_capacitance = 0.1e-15@u_F
parasitic_capacitance = 0.1e-15@u_F
parasitic_inductance = 1@u_mH
tmdc_resistance = 100@u_kΩ
title = 'Monolayer MoS2'

# initial setup parameters
external_resistance = 10@u_MΩ
pulse_frequency = 80e6
chopper_frequency = 50e3

# simulation parameters
# steps per period
resolution = 500
# number of periods
length = 10
# ambient temp (does not matter)
t = 20
# pulse frequency (False) or chopper frequency (True)
fmode = False
```

```python
# circuit initialization

def init_circuit():
    global r_ext, c_p, c_j1, r_j1, r_m, c_j2, r_j2, l_p, i_photo

    ml_mos2 = Circuit(title)
    ml_mos2

    r_ext = ml_mos2.R('_ext', ml_mos2.gnd, 'out', external_resistance)
    c_p = ml_mos2.C('_p', ml_mos2.gnd, 'out', parasitic_capacitance)
    c_j1 = ml_mos2.C('_j1', 'out', 'ml_contact_1', junction_capacitance)
    r_j1 = ml_mos2.R('_j1', 'out', 'ml_contact_1', junction_resistance)
    r_m = ml_mos2.R('_tmdc', 'ml_contact_1', 'ml_contact_2', tmdc_resistance)
    c_j2 = ml_mos2.C('_j2', 'ml_contact_2', ml_mos2.gnd, junction_capacitance)
    r_j2 = ml_mos2.R('_j2', 'ml_contact_2', ml_mos2.gnd, junction_resistance)
    l_p = ml_mos2.L('_p', ml_mos2.gnd, 'out', parasitic_inductance)

    if not fmode:
        i_photo = ml_mos2.PulseCurrentSource('_pulse', 'ml_contact_1', 'out', initial_value=0, pulsed_value=1@u_uA,
                                         pulse_width=1/(2*pulse_frequency)@u_s, period=1/(pulse_frequency)@u_s)
        print('PULSE INCLUDED')
        print(f'pulse freq: {i_photo.frequency}\n')
        chop_toggle = False

    elif fmode:
        i_photo = ml_mos2.PulseCurrentSource('_chopper', 'ml_contact_1', 'out', initial_value=0, pulsed_value=1@u_uA,
                                         pulse_width=1/(4*chopper_frequency)@u_s, period=1/(2*chopper_frequency)@u_s)
        print('PULSE EXCLUDED')
        print(f'chopper freq: {i_photo.frequency}\n')
        chop_toggle = True
    
    print(ml_mos2)

    return ml_mos2.simulator(temperature=t, nominal_temperature=t)

simulator = init_circuit()
```

```python
# transient analysis at initial values

analysis = simulator.transient(step_time=i_photo.period/resolution, end_time=i_photo.period*length)

fig, ax = plt.subplots(figsize=(16, 10))
ax.plot(np.array(analysis.out.abscissa), np.array(analysis.out), 'x-')
ax.set_xlabel('time [s]')
ax.set_ylabel('voltage at $R_{ext}$ [V]')
```

```python
# frequency sweep at selected param values
# list values
junction_r_vals = [i@u_Ω for i in np.logspace(5, 6, 1)] # default: 10^5 Ohm
junction_c_vals = [i@u_F for i in np.logspace(-18, -12, 7)] # default: 10^-16 F
tmdc_r_vals = [i@u_Ω for i in np.logspace(5, 6, 1)] # default: 10^5 Ohm
parasitic_c_vals = [i@u_F for i in np.logspace(-18, -5, 2)] # default: 10^-16 F
parasitic_ind_vals = [i@u_H for i in np.logspace(-3, -5, 1)] # default: 10^-3 H
# frequency sweep
fmin = 1e1
fmax = 1e7
fresolution = 100

iter_count = len(junction_r_vals)*len(junction_c_vals)*len(tmdc_r_vals)*len(parasitic_c_vals)*len(parasitic_ind_vals)
print(f'iterations: {iter_count}')

start = datetime.utcnow()
chop_sweep_result = sweep_param('frequency', (fmin, fmax), fresolution, i_photo, simulator, [i_photo], choppermode=fmode)
stop = datetime.utcnow()
expect = (stop-start)*iter_count
print(f'expected time: {expect}')
```

```python
results = []
params_data = []

start = datetime.utcnow()
print(f'started at {start}, projected finish: {start+expect}')

for rj in junction_r_vals:
    r_j1.resistance = rj
    r_j2.resistance = rj
    for rtmdc in tmdc_r_vals:
        r_m.resistance = rtmdc
        for cp in parasitic_c_vals:
            c_p.capacitance = cp
            for cj in junction_c_vals:
                c_j1.capacitance = cj
                c_j2.capacitance = cj
                for indp in parasitic_ind_vals:
                    l_p.inductance = indp
                    chop_sweep_result = sweep_param('frequency', (fmin, fmax), fresolution, i_photo, simulator, [i_photo], 
                                                    choppermode=fmode, output=False)
                    results.append(chop_sweep_result)
                    params_data.append([rj, cj, rtmdc, cp, indp])
                    
print(f'took {datetime.utcnow()-start} for {iter_count} iterations')
```

```python
%matplotlib auto

max_amps_collection = np.zeros((len(results), len(results[0]), 2))

for ind, res in enumerate(results):
    for ind2, sim in enumerate(res):
        data, freq = sim
        max_amps_collection[ind, ind2, 0] = np.array(data.out).max()
        max_amps_collection[ind, ind2, 1] = freq

fig2, ax2 = plt.subplots(figsize=(16, 16))
if not fmode:
    ax2.set_xlabel('laser frequency', fontsize=18)
else:
    ax2.set_xlabel('chopper frequency', fontsize=18)
    max_amps_collection[:, :, 1] *= 0.5

save_array = np.zeros((8, len(params_data)), dtype=float)
ind_count = 3

for ind, data in enumerate(max_amps_collection):
    if ind % 2 == 1:
        ax2.plot(data[:, 1], data[:, 0], 'o-', label=ind_count, markersize=3)
        save_array[:5, ind] = params_data[ind]
        ind_count += 1
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_ylabel('Voltage at out-gnd node', fontsize=18)
ax2.legend(fontsize=18)
ax2.grid()

save_array[5,:] = float(external_resistance)
save_array[6,:] = float(pulse_frequency)
if fmode:
    save_array[7,:] = 1
else:
    save_array[7,:] = 0
save_frame = pd.DataFrame({'Rj': save_array[0, :], 'Cj': save_array[1, :], 'Rtmdc': save_array[2, :],
                          'Cp': save_array[3, :], 'Lp': save_array[4, :], 'Rex': save_array[5, :], 
                          'fpulse': save_array[6, :], 'mode': save_array[7, :]})

print(save_frame)
```

```python
save_id = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
fig2.savefig(f'.\sim_out\{save_id}.pdf')
save_frame.to_csv(f'.\sim_out\{save_id}.csv')
save_frame.to_excel(f'.\sim_out\{save_id}.xlsx')
```

```python
# param sweep at selected frequencies
frequencies = [80e6]

simulator = init_circuit()

params = [c_p]
pmin = 1e-12
pmax = 1e-5
presolution = 200
ptype = 'capacitance'

results = []

for fr in frequencies:
    i_photo.pulse_width = 1/(2*fr)
    i_photo.period = 1/(fr)
    results.append(sweep_param(ptype, (pmin, pmax), presolution, i_photo, simulator, params))
```

```python
%matplotlib auto
max_amps_collection = np.zeros((len(results), len(results[0]), 2))
params_data = []

for ind, res in enumerate(results):
    for ind2, sim in enumerate(res):
        data, xval = sim
        max_amps_collection[ind, ind2, 0] = np.array(data.out).max()
        max_amps_collection[ind, ind2, 1] = xval
    params_data.append([r_j1.resistance, c_j1.capacitance, r_m.resistance, c_p.capacitance, l_p.inductance])

save_array = np.zeros((8, len(params_data)), dtype=float)     
        
fig4, ax4 = plt.subplots(figsize=(16, 9))
for ind, data in enumerate(max_amps_collection):
    ax4.plot(max_amps_collection[ind, :, 1], max_amps_collection[ind, :, 0], 'o-', label=f'{frequencies[ind]} Hz', 
            markersize=3)
    save_array[:5, ind] = params_data[ind]
    save_array[6, ind] = frequencies[ind]
ax4.set_xscale('log')
# ax4.set_yscale('log')
ax4.set_ylabel('Voltage at out-gnd node', fontsize=18)
ax4.set_xlabel(params[0].name, fontsize=18)
ax4.set_title(f'{params[0].name} sweep', fontsize=18)
ax4.legend()

save_array[5,:] = float(external_resistance)
if fmode:
    save_array[7,:] = 1
else:
    save_array[7,:] = 0
save_frame = pd.DataFrame({'Rj': save_array[0, :], 'Cj': save_array[1, :], 'Rtmdc': save_array[2, :],
                          'Cp': save_array[3, :], 'Lp': save_array[4, :], 'Rex': save_array[5, :], 
                          'fpulse': save_array[6, :], 'mode': save_array[7, :]})

print(save_frame)
```

```python
save_id = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
fig4.savefig(f'.\sim_out\{save_id}.pdf')
save_frame.to_csv(f'.\sim_out\{save_id}.csv')
save_frame.to_excel(f'.\sim_out\{save_id}.xlsx')
```
